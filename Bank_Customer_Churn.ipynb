```python
# Dropping Irrelevant Features
df=df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
df.head()
```

```python
#Encoding categorical data
df = pd.get_dummies(df,drop_first=True)
```

```python
df.head()
```

```python
df['Exited'].value_counts()
```

```python
#store the dataset in X and Y
X = df.drop('Exited', axis=1)
Y = df['Exited']
```

```python
#handling Imbalanced Data with SMOTE
from imblearn.over_sampling import SMOTE
```

```python
X_res,Y_res = SMOTE().fit_resample(X,Y)
```

```python
Y_res.value_counts()
```

```python
#Splitting the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X_res,Y_res,test_size=0.2,random_state=42)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
my_model = LogisticRegression()
```

```python
my_model.fit(X_train,Y_train)
```

```python
y_prediction = my_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_prediction,Y_test)
```

```python
#SVC- Support Vector Classifire.
from sklearn import svm
```

```python
#create a model
my_model2=svm.SVC()
```

```python
#fit data into the model
my_model2.fit(X_train,Y_train)
```

```python
#perform prediction
y_pred2 = my_model2.predict(X_test)
```

```python
#check model accuracy
accuracy_score(Y_test,y_pred2)
```

```python
#K-Neighbors Clssifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Crete Model
knn = KNeighborsClassifier()
```

```python
#Fit data into the model
knn.fit(X_train,Y_train)
```

```python
#Test the model
y_pred3 = knn.predict(X_test)
```

```python
#Check ccuarcy score
accuracy_score(Y_test,y_pred3)
```

```python
#Decision Tree Clssifire
from sklearn.tree import DecisionTreeClassifier
```

```python
#Creat Model
my_dt = DecisionTreeClassifier()
```

```python
#fit the data into the model
my_dt.fit(X_train,Y_train)
```

```python
#test the model - prediction
y_pred4 = my_dt.predict(X_test)
```

```python
#acuuracy check
accuracy_score(Y_test,y_pred4)
```

```python
#now make new dataframe for all modeles
final_data = pd.DataFrame({'Models':['LR','SVC','KNN','DT'],
                           'ACCURACY':[accuracy_score(Y_test,y_prediction),
                                       accuracy_score(Y_test,y_pred2),
                                       accuracy_score(Y_test,y_pred3),
                                       accuracy_score(Y_test,y_pred4)]})
```
